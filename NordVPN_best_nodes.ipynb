Reading all the CA/TLS from nordvpn

```python
from os import walk

count = 0
f = []

for (dirpath, dirnames, filenames) in walk("CA_and_TLS_auth_certificates"):
    for file in filenames:
        count+=1
        if count % 2 == 0:
            f.append(file.replace("_ca.crt","").\
                     replace("_tls.key","").\
                     replace("_", "."))
```

Checking the round-trip-time to each nordvpn instance

```python
import subprocess

result=[]
for file in f:
    try:
        result.append(subprocess.check_output(["ping","-c","1","-t","1",file], universal_newlines=True).replace("\n",";"))
    except:
        continue
```

Filtering results

```python
filtered_results= dict()

for r in result:
    filtered_results[r.split("PING ")[1].split(" ")[0]] = float(r.split("PING ")[1].split(" ")[10].replace("time=",""))
```

Sorting everything 

```python
import operator
sorted(filtered_results.items(), key=operator.itemgetter(1))
```
